## Wind Analysis: Polar Plots for Wind Speed and Direction

### Objective

To identify trends and significant wind events by analyzing the distribution of wind speed (`WS`) and wind direction (`WD`), along with the variability in wind direction. Polar plots are particularly useful for visualizing the relationship between wind speed and direction.

### Steps

1. **Preparation**:
   - Extract the wind speed (`WS`), wind gust (`WSgust`), and wind direction (`WD`) data from the dataset.
   - Ensure that wind direction is properly represented as an angular measurement in degrees.

2. **Wind Speed and Direction Polar Plot**:
   - **Polar Plot Basics**:
     - The plot will use wind direction (`WD`) as the angle on the polar axis, and wind speed (`WS`) or wind gust (`WSgust`) as the radius.
   - **Interpretation**:
     - The plot will show how wind speed varies with direction, helping to identify dominant wind directions and any significant wind events (e.g., high wind speeds from a particular direction).

3. **Wind Direction Variability**:
   - **Objective**:
     - Assess how variable the wind direction tends to be, which can be crucial for understanding the stability of wind patterns and potential impacts on solar installations or other projects.
   - **Approach**:
     - Calculate the variability or standard deviation of wind direction within specific time intervals (e.g., hourly, daily) and visualize it on a polar plot.



```python

# Convert wind direction to radians for polar plot
df['WD_radians'] = np.deg2rad(df['WD'])

# Plot wind speed and direction on a polar plot
plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(df['WD_radians'], df['WS'], alpha=0.75, cmap='viridis')
ax.set_theta_direction(-1)  # Clockwise direction
ax.set_theta_offset(np.pi / 2.0)  # Zero degrees at the top
plt.title('Polar Plot of Wind Speed and Direction')
plt.show()

# Plot wind gust and direction on a polar plot
plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)
ax.scatter(df['WD_radians'], df['WSgust'], alpha=0.75, cmap='plasma')
ax.set_theta_direction(-1)  # Clockwise direction
ax.set_theta_offset(np.pi / 2.0)  # Zero degrees at the top
plt.title('Polar Plot of Wind Gust and Direction')
plt.show()

# Optional: Assess wind direction variability
# Calculate standard deviation of wind direction for time intervals (e.g., hourly)
wind_direction_variability = df.groupby(df.index.hour)['WD'].std()

# Plot wind direction variability on a polar plot
plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)
ax.plot(np.deg2rad(wind_direction_variability.index * 15), wind_direction_variability, 'b-')
ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi / 2.0)
plt.title('Wind Direction Variability Throughout the Day')
plt.show()
```

   - Add regression lines to understand the trend.

3. **Heatmaps**:
   - Create heatmaps to visualize how the correlation between `RH` and temperature/solar radiation varies over different times of the day or during different months.

## Exploring Complex Relationships with Bubble Charts

### Objective

The aim is to use bubble charts to explore the relationships between `GHI` (Global Horizontal Irradiance), `Tamb` (Ambient Temperature), and `WS` (Wind Speed), with the size of the bubbles representing an additional variable like `RH` (Relative Humidity) or `BP` (Barometric Pressure). Bubble charts can effectively visualize multivariate relationships, helping to uncover patterns or correlations that might not be immediately apparent.

### Approach

1. **Bubble Chart 1: GHI vs. Tamb vs. WS, Bubble Size Representing RH**:
   - This chart will help us understand how `GHI` and `Tamb` interact with `WS`, while the bubble size indicates the level of relative humidity.

2. **Bubble Chart 2: GHI vs. Tamb vs. WS, Bubble Size Representing BP**:
   - This variation of the bubble chart uses `BP` (Barometric Pressure) to explore how atmospheric pressure influences the relationship between solar irradiance, temperature, and wind speed.


```python

# Bubble Chart 1: GHI vs. Tamb vs. WS, Bubble Size = RH
plt.figure(figsize=(12, 8))
plt.scatter(df['Tamb'], df['GHI'], s=df['RH']*10, c=df['WS'], alpha=0.5, cmap='viridis')
plt.colorbar(label='Wind Speed (WS)')
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Global Horizontal Irradiance (GHI)')
plt.title('Bubble Chart: GHI vs. Tamb vs. WS, Bubble Size = RH')
plt.show()

# Bubble Chart 2: GHI vs. Tamb vs. WS, Bubble Size = BP
plt.figure(figsize=(12, 8))
plt.scatter(df['Tamb'], df['GHI'], s=df['BP']*0.1, c=df['WS'], alpha=0.5, cmap='coolwarm')
plt.colorbar(label='Wind Speed (WS)')
plt.xlabel('Ambient Temperature (Tamb)')
plt.ylabel('Global Horizontal Irradiance (GHI)')
plt.title('Bubble Chart: GHI vs. Tamb vs. WS, Bubble Size = BP')
plt.show()
```
